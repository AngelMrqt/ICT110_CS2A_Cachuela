Exercise 2

Write a program to prompt the user for hours and rate per hour to compute gross pay.

Enter Hours: 35
Enter Rate: 2.75
Pay: 96.25

```python
hours = float(input('Enter hours worked: '))
rate = float(input('Enter rate per hour: '))

pay = hours * float(rate)

print('Pay: ', pay)
```
